```python
# drop FIPS columns
df_td=df_td.drop(columns=['State FIPS', 'County FIPS'])
```

```python
# consistancy check
df_td.describe()
```

```python
# check missing values
df_td.isnull().sum()
```

```python
# Null values are expected for State Postal Code and County Name for the National level rows
```

```python
df_nan=df_td[df_td['Number of Trips'].isnull()==True]
```

```python
df_nan
```

```python
# delete rows with no data
```

```python
df_td.shape
```

```python
df_td_clean = df_td[df_td['Number of Trips'].isnull()==False]
```

```python
df_td_clean.shape
```

```python
# Check for duplicates
df_dups = df_td_clean[df_td_clean.duplicated()]
```

```python
df_dups
```

```python
# no duplicates
```

```python
df_td_clean.describe()
```

```python
# why does 'Population Not Staying Home' have a minimum less than 0?
```

```python
df_td_clean.loc[df_td_clean['Population Not Staying at Home'] < 0]
```

```python
# remove all records with values less than 0
df_td_clean2=df_td_clean[df_td_clean['Population Not Staying at Home'] >= 0]
```

```python
df_td_clean2.shape
```

```python
# Check for mixed types
for col in df_td_clean2.columns.tolist():
  weird = (df_td_clean2[[col]].map(type) != df_td_clean2[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_td_clean2[weird]) > 0:
    print (col)
```

```python
# these mixed values are due to the NaN values in the National column
```

```python
# export dataframe
df_td_clean2.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Trips_by_Distance_cleaned.csv'))
```
