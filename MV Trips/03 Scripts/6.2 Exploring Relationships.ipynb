```python
sub = trips[['Population Staying at Home', 'Population Not Staying at Home', 'Number of Trips']]
```

```python
#Create a heatmap
plt.matshow(sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# import data
tripsN = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Trips_by_Distance_national.csv'))
```

```python
sns.histplot(tripsN['Population Not Staying at Home'], bins = 20, kde = True)
plt.show()
```

```python
tripsN.columns
```

```python
tripsN.head()
```

```python
# define category
tripsN.loc[tripsN['Population Not Staying at Home'] < 250000000, 'Travelers category'] = 'low travelers'
tripsN.loc[(tripsN['Population Not Staying at Home'] >= 250000000) & (tripsN['Population Not Staying at Home'] < 265000000), 'Travelers category'] = 'medium travelers'
tripsN.loc[tripsN['Population Not Staying at Home'] >= 265000000, 'Travelers category'] = 'high travelers'
```

```python
tripsN['Travelers category'].value_counts(dropna = False)
```

```python
# create categorical plot of population not staying home vs number of trips
sns.set(style="ticks")
g = sns.catplot(x="Number of Trips", y="Population Not Staying at Home", hue="Travelers category", data=tripsN)
```

```python
plt.show()
```
